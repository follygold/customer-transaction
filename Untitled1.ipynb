```python
import pandas as pd

df = pd.read_excel('Book1.xlsx')

display(df.describe())
df.head()
```

```python
display(df.info())
df.nunique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(df.shape)
qual_col = [col for col in df.columns if df[col].dtype == 'object']
quan_col = [col for col in df.columns if df[col].dtype != 'object']
print('qual_col', qual_col)
print('quan_col', quan_col)
```

```python
fig, ax = plt.subplots(3, 2,figsize=(60,50))
ax = ax.flatten()
i = 0 
sns.set(font_scale = 2)

for col in qual_col:
    if col not in ['name', 'job_title']:
        sns.countplot(x= col, ax = ax[i], data = df)
        ax[i].set_title('count of customer by '+ col, fontsize = 50)
        ax[i].tick_params(labelsize = 30)
        i += 1
```

<br> More female buys a bike than males but the diffrence is little </br>
<br> Most bike owners or customers are from manufacturing, financial service, and health industry with the least from agriculture and communication industry </br>
<br> most bike owners are niether affluent nor have a high net worth </br>
<br> all bike owners are alive <br>
<br> possesion of a have insignificant effect to owning a bike. <br>

```python
df['year'] = df['DOB'].dt.year

fig, ax = plt.subplots(2, 1, figsize = (20, 12))
df['year'].value_counts().sort_index().plot(ax = ax[0]) ;
ax[0].set_title('customer plot by year of birth')
df['age'].value_counts().sort_index().plot(ax = ax[1]) ;
ax[1].set_title('customer plot by age') ;
```

a lot of bike owners were born around 1972 o 1980 and most customers are between 20 to 67 years from the above diagram

```python
fig, ax = plt.subplots(2, 1, figsize = (20, 12))
df['past_3_years_bike_related_purchases'].value_counts().sort_index().plot(ax = ax[0]) ;
ax[0].set_title('customer plot by past_3_years_bike_related_purchases')
df['tenure'].value_counts().sort_index().plot(ax = ax[1]) ;
ax[1].set_title('customer plot by tenure') ;
```

```python
fig, ax = plt.subplots(2, 2,figsize=(60,45))
ax = ax.flatten()
i = 0 
sns.set(font_scale = 2)

for col in quan_col:
    if col not in ['customer_id', 'DOB', 'owns_car']:
        sns.distplot(df[col].values, ax = ax[i])
        ax[i].set_title('count of customer by '+ col, fontsize = 50)
        ax[i].tick_params(labelsize = 30)
        i += 1
```
