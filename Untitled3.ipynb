```python
display(df_clist.head())
quan_col = ['past_3_years_bike_related_purchases', 
            'tenure', 'postcode', 'property_valuation', 'Rank', 'Value']
qual_col = [col for col in df_clist.columns if col not in quan_col + ['DOB']]
df_clist[quan_col] = df_clist[quan_col].astype('float32')
df_clist['DOB'] = pd.to_datetime(df_clist['DOB'], format = '%Y/%m/%d', errors = 'coerce')
```

## Analysis

```python
fig,  ax = plt.subplots(11,1, figsize = (60, 200))
ax = ax.flatten()
i = 0
for col in qual_col:
    #removing null values before plotting
    data = df_clist[[col]].dropna()
    sns.countplot(x = col, data = data, ax = ax[i])
    ax[i].set_title('count of customer by '+ col, fontsize = 50)
    ax[i].tick_params(labelsize = 30)
    i += 1
    
```

Most job titles have one customer lets check job titles with up to 5 customers

#### analysing job titles using specific key word in job title

```python
job_tls = list(df_clist['job_title'].dropna().unique())

col = {}
for tls in job_tls:
    #print(tl)
    for tl in tls.split(' '):
        if tl.upper() in col:
            col[tl.upper()] += 1
        else:
            col[tl.upper()] =1
```

```python
dat = pd.Series(col)
dat = pd.DataFrame(dat)#.reset_index()
dat.columns = ['Count']
dath = dat[dat['Count'] > 5]
datl = dat[dat['Count'] < 5]

sns.set(font_scale = 1)
plt.figure(figsize =(20, 9)) ;
sns.barplot( data = dath, x = dath.index, y = dath['Count']) ;
plt.title('job title key words with highest customers', fontsize = 15) ;
plt.show()

plt.figure(figsize =(20, 9)) ;
sns.barplot( data = datl, x = datl.index, y = datl['Count']) ;
plt.title('job title key words with lowest customers', fontsize = 15) ;
```
